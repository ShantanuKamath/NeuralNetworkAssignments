```python
import glob
import json
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
with open('../logs_2/3_layer_mlp_30_1e-05_full.json') as f:
    data = json.load(f)
    plt.plot(data['epoch'], data['train_loss'], label="train error")
    plt.ylabel('Mean Squared Error')
    plt.xlabel('Epochs')
    plt.legend()
    plt.savefig('Partb_1a', dpi=300)
```

```python
with open('../logs_2/3_layer_mlp_30_1e-05_full.json') as f:
    data = json.load(f)
    plt.plot(data['epoch'], data['test_loss'], label="test error")
    plt.ylabel('Mean Squared Error')
    plt.xlabel('Epochs')
    plt.legend()
    plt.savefig('Partb_1b', dpi=300)
```

```python
files = {1: glob.glob('../logs_2/3_layer_mlp_30_0.001_split_*.json'),
         2: glob.glob('../logs_2/3_layer_mlp_30_0.0005_split_*.json'),
         3: glob.glob('../logs_2/3_layer_mlp_30_0.0001_split_*.json'),
         4: glob.glob('../logs_2/3_layer_mlp_30_5e-05_split_*.json'),
         5: glob.glob('../logs_2/3_layer_mlp_30_1e-05_split_*.json')}
for i in range(1,6):
    best_acc = 1000000
    best_file = ''
    for f in files[i]:
        with open(f) as a:
            data = json.load(a)
            if (np.min(data['val_acc']) < best_acc):
                best_acc = np.min(data['val_acc'])
                best_file = data
    files[i] = best_file


lr = ['0.001', '0.0005', '0.0001', '5e-5', '1e-5']
for i in range(1,6):
    data = files[i]
    plt.plot(data['epoch'], data['val_loss'], label=lr[i-1])
plt.ylabel('Mean Squared Error')
plt.xlabel('Epochs')
plt.legend()
plt.savefig('Partb_2b', dpi=300)
```

```python
with open('../logs_2/3_layer_mlp_30_1e-05_full.json') as f:
    data = json.load(f)
    plt.plot(data['epoch'], data['test_loss'], label="test error")
    plt.ylabel('Mean Squared Error')
    plt.xlabel('Epochs')
    plt.legend()
    plt.savefig('Partb_2c', dpi=300)
```

```python
files = {1: glob.glob('../logs_2/3_layer_mlp_20_1e-05_split_*.json'),
         2: glob.glob('../logs_2/3_layer_mlp_30_1e-05_split_*.json'),
         3: glob.glob('../logs_2/3_layer_mlp_40_1e-05_split_*.json'),
         4: glob.glob('../logs_2/3_layer_mlp_50_1e-05_split_*.json'),
         5: glob.glob('../logs_2/3_layer_mlp_60_1e-05_split_*.json')}
for i in range(1,6):
    best_acc = 1000000
    best_file = ''
    for f in files[i]:
        with open(f) as a:
            data = json.load(a)
            if (np.min(data['_acc'] < best_acc):
                best_acc = data['test_acc']
                best_file = data
    files[i] = best_file

bs = ['20', '30', '40', '50', '60']
for i in range(1,6):
    data = files[i]
    plt.plot(data['epoch'], data['train_loss'], label=bs[i-1])
plt.ylabel('Mean Squared Error')
plt.xlabel('Epochs')
plt.legend()
plt.savefig('Partb_3a', dpi=300)
```
